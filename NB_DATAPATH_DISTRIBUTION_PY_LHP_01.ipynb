```python
from notebookutils import mssparkutils
ruta = "Files/gold/dim_brokers"
if mssparkutils.fs.exists(ruta):
    mssparkutils.fs.rm(ruta, True)
try:
    df = spark.read.table("`brokers csv`")
    df.write.mode("overwrite").saveAsTable("dim_brokers")

    # Escribirla en formato delta en Files/gold
    df.write.format("delta") \
    .option("overwriteSchema", "true") \
    .mode("overwrite") \
    .save("Files/gold/dim_brokers")
    print("✅ Data Gold guardada")

    spark.sql("DROP TABLE `brokers csv`")
    print("✅ Tabla Eliminada")
except Exception as e:
    print(f"Error al guardar: {str(e)}")
```
