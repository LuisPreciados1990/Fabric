```python
import pyspark.sql.functions as F
from pyspark.sql.types import *
from notebookutils import mssparkutils

schema_brokers = StructType([
    StructField("id_broker", IntegerType(), False),
    StructField("name", StringType(), True),
    StructField("region", StringType(), True),
    StructField("email", StringType(), True)
])

try:
    df_brokers = spark.read.format("csv") \
        .option("header", "true") \
        .schema(schema_brokers) \
        .load("Files/broze/brokers.csv")

    df_brokers.coalesce(1).write \
    .option("header", True) \
    .mode("overwrite") \
    .csv("Files/silver/temp_brokers")

    files = mssparkutils.fs.ls("Files/silver/temp_brokers")
    part_file = [f.path for f in files if "part-" in f.name][0]
    final_path = "Files/silver/brokers.csv"

    mssparkutils.fs.mv(part_file, final_path)
    mssparkutils.fs.rm("Files/silver/temp_brokers", True)
    print("✅ Esquemas creado y guardado")
except Exception as e:
    print(f"Error al leer archivo: {str(e)}")
```

```python
schema_properties = StructType([
    StructField("id_property", IntegerType(), False),
    StructField("id_project", IntegerType(), False),
    StructField("type", StringType(), True),
    StructField("size_m2", DoubleType(), True),
    StructField("rooms", IntegerType(), True),
    StructField("bathroms", IntegerType(), True),
    StructField("price", DoubleType(), True),
    StructField("status", StringType(), True)
])


try:
    df_properties = spark.read.format("csv") \
        .option("header", "true") \
        .schema(schema_properties) \
        .load("Files/broze/properties.csv")

    df_project_rename = df_projects \
        .withColumnRenamed("name", "name_project") \
        .withColumnRenamed("city", "city_project") \
        .withColumnRenamed("region", "region_project") \
        .withColumnRenamed("status", "status_project") \
        .withColumnRenamed("release_date", "release_date_project")

    df_property_rename = df_properties \
        .withColumnRenamed("price", "price_property")\
        .withColumnRenamed("status", "status_property")

    df_properties_desnormalized = df_property_rename.join(
        df_project_rename,
        on="id_project", 
        how="left"
    ).select(
        "id_property",
        "id_project",
        "type",
        "size_m2",
        "rooms",
        "bathroms",
        "price_property",
        "status_property",
        "name_project",
        "city_project",
        "region_project",
        "release_date_project",
        "status_project"
    )
    df_properties_desnormalized.coalesce(1).write \
    .option("header", True) \
    .mode("overwrite") \
    .csv("Files/silver/temp_properties")

    files = mssparkutils.fs.ls("Files/silver/temp_properties")
    part_file = [f.path for f in files if "part-" in f.name][0]
    final_path = "Files/silver/properties.csv"

    mssparkutils.fs.mv(part_file, final_path)
    mssparkutils.fs.rm("Files/silver/temp_properties", True)
    print("✅ Esquemas creado y guardado")
except Exception as e:
    print(f"Error al leer archivo: {str(e)}")
```
